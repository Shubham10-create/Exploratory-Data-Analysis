## Covid Live Tracking For India

```python
import pandas as pd

df = pd.read_json("https://www.mohfw.gov.in/data/datanew.json")
df
```

```python
df = df.iloc[:-1,1:]
```

```python
df
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
total_states = np.arange(len(df['state_name']))
```

```python
total_states
```

```python
df['active']
```

```python
max(df['positive'])
```

## Total Positive Cases Based on States in Maharashtra

```python
#Total positive cases
from matplotlib.pyplot import figure
figure(num=None, figsize=(10,7), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['positive'], align='center', alpha=0.5, color=(1,0,0))
plt.yticks(total_states, df['state_name']) #putting state names
plt.xlim(1,max(df['positive'])+100) #the x limit
plt.xlabel("Positive Cases")
plt.title("Covid Cases")
plt.show()
```

## Active Cases Based On States in India 

```python
#Active cases in India
from matplotlib.pyplot import figure
figure(num=None, figsize=(10,7), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['active'], align='center', alpha=0.5, color=(1,0.5,0))
plt.yticks(total_states, df['state_name']) #putting state names
plt.xlim(1,max(df['active'])+10) #the x limit
plt.xlabel("Active Cases")
plt.title("Active Covid Cases")
plt.show()
```

## Total Deaths In India State Wise

```python
#Total Deaths in Indian States
from matplotlib.pyplot import figure
figure(num=None, figsize=(10,7), dpi=80, facecolor='w', edgecolor='k')
plt.barh(total_states, df['death'], align='center', alpha=0.5, color=(1,0,0))
plt.yticks(total_states, df['state_name']) #putting state names
plt.xlim(1,max(df['death'])+10) #the x limit
plt.xlabel("Total deaths")
plt.title("Deaths due to Covid")
plt.show()
```
